```python
import pandas as pd
import numpy as np
```

```python
measles_data = pd.read_csv('../data/Weekly_measles_cases_1948_to_1967.csv')
print(measles_data.columns)
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
```

```python
print('Number of data points: {}'.format(measles_data.shape[0]))
```

```python
max_date = measles_data['Year'][np.argmax(measles_data['Weekly measles cases'])]
max_occ = measles_data['Weekly measles cases'].max()

min_date = measles_data['Year'][np.argmin(measles_data['Weekly measles cases'])]
min_occ = measles_data['Weekly measles cases'].min()


print('Highest number of weekly cases {} occured in {}.'.format(max_occ, max_date))
print('Lowest number of weekly cases {} occured in {}.'.format(min_occ, min_date))
```

```python
plt.plot(measles_data['Weekly measles cases'])
plt.title('Number of weekly measles cases', size=16)
plt.xticks(np.arange(measles_data.shape[0])[0::52], 
           [int(x) for x in measles_data['Year'][0::52]], rotation=45)
plt.xlabel('Time')
plt.ylabel('Number of weekly cases')
plt.show()
```

```python
years = [x for x in measles_data['Year'][::52]]
yearly_data = []
for x in range(len(years)):
    yearly_data.append(measles_data['Weekly measles cases'][x*52:(x+1)*52])
```

```python
np.arange(1,52,4.5)
```

```python
from itertools import product
```

```python
combo = list(product(['k','r','b','g','y','m','c'],['-','--','-.']))
colors = [''.join(x) for x in combo]
```

```python
len(colors)
```

```python
plt.figure(figsize=(8,8))
plt.title('Weekly Measles cases yearly comparisson')
for i ,year in enumerate(yearly_data):
    plt.plot(range(len(year)), year,colors[i],label=str(int(years[i])))
plt.xticks(np.arange(0,52,4.3),['January','February','March','April','May',
                                'June','July','August','September','October',
                                'November','December'],rotation=45)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1), prop={'size':10})
plt.xlabel('Month')
plt.ylabel('Weekly cases')
plt.margins(0.02)
plt.show()
```
